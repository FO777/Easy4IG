```python
#single
import requests
from random import randint
import json
from time import sleep
from InstagramAPI import InstagramAPI
from requests.packages.urllib3.exceptions import InsecureRequestWarning
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

def sendMessage(self, target_user, msgText): #target user must be a string also it is an identity token and not their username
                                            #example str(12188447211)
    target_user = '[[{}]]'.format(','.join([target_user]))
    url = 'direct_v2/threads/broadcast/text/'
    data = {
    'text': msgText,
    '_uuid': self.uuid,
    '_csrftoken': self.token,
    'recipient_users': target_user,
    '_uid': self.username_id,
    'action': 'send_item',
    'client_context': self.generateUUID(True)}
    return self.SendRequest(url, data)

def like(self, media): #like, needs mediaID
    data = json.dumps({'_uuid': self.uuid,
                       '_uid': self.username_id,
                       '_csrftoken': self.token,
                       'media_id': media})
    return self.SendRequest('media/' + str(media) + '/like/', self.generateSignature(data))

def comment(self, mediaId, commentText): #comment, needs mediaID
    data = json.dumps({'_uuid': self.uuid,
                       '_uid': self.username_id,
                       '_csrftoken': self.token,
                       'comment_text': commentText})
    return self.SendRequest('media/' + str(mediaId) + '/comment/', self.generateSignature(data))

numAccts = 1

u = []
p = []
for i in range(numAccts):
    username = input("what is the username of account " +str(i+1)+ ": ")
    password = input("what is the password of account " +str(i+1)+ ": ")
    u.append(username)
    p.append(password)
    
    c, m, h, r, f = [], [], [], [], []
    print("if you do not want messages to be active leave the answer blank")
    for i in range(numAccts):
        c.append([])
        m.append([])
        h.append([])
        f.append([])
        r.append(InstagramAPI(u[i], p[i]))
        hashtag = input("pick a target hashtag for account "+str(u[i])+": ")
        h[i].append(hashtag)
        messages = input("direct message for account "+str(u[i])+": ")
        m[i].append(messages)
    print("if you do not want comments to be active leave the answer blank")
    for i in range(len(c)):
        for n in range(5):
            comments = input("comment "+str(n+1)+" for account "+str(u[i])+":")
            c[i].append(comments)

def running():
    print('running... wait at least 15 mins while the bot turns on and does its checks')
    for i in range(numAccts):
        r[i].login()
        r[i].getTotalSelfFollowers()
        r[i].follows = []
        for n in range(len(r[i].follows)):
            f[i].append(r[i].follows[n]['pk'])
        sleep(60)
    print("bot has logged in, it will start its work now, check your account in an hour and look at your previously liked posts in settings>account>posts you've liked")
    while True:
        for num in range(10):
            for i in range(numAccts):
                if m[i] != ['']:
                    r[i].check = r[i].getTotalSelfFollowers()
                    for n in range(len(r[i].check)):
                        if r[i].check[n]['pk'] not in f[i]:
                            sendMessage(r[i], str(r[i].check[n]['pk']), m[i][0])
                            f[i].append(r[i].check[n]['pk'])
                            sleep(30)
                r[i].getHashtagFeed(h[i][0])
                r[i].count = r[i].LastJson
                for row in range(10):
                    user = r[i].count['items'][row]['caption']['media_id']
                    like(r[i], user)
                    sleep(randint(10, 15))
                    cc = c[i][randint(0,4)]
                    if cc != '':
                        comment(r[i], user, cc)
                    sleep(randint(10, 15))
            sleep(8640/numAccts)
        for i in range(numAccts):
            r[i].login()
            sleep(60)
        
running()
```
